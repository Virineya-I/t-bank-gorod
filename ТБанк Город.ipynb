# Т-Банк: использование сервисов Города

## Описание проекта

  Т-Город — это группа сервисов, которые помогают пользователям решать ежедневные
задачи. Например, есть сервисы, которые позволяют купить билеты в кино или заказать
продукты. Одна из задач, которая стоит перед аналитиками, — понять, чем клиенты
Города отличаются от неклиентов. Дана база данных клиентов и их характеристики,
а также флаг, покупал ли человек хотя бы 1 раз в Городе, и статистика покупок через
основные сервисы. Основными считаются три сервиса: Топливо, Продукты и Афиша.  

  **Топливо** — сервис, который позволяет заправить машину, не выходя из нее, а оплатив
заправку дистанционно.       
  **Продукты** — сервис по доставке продуктов из различных магазинов. Например, можно
заказать домой продукты из Вкусвилла, Пятёрочки, Перекрёстка, Ленты.     
  **Афиша** — сервис по покупке билетов на различные мероприятия: в кино, театры и на
концерты.    

  В данных вам даны скоры принадлежности к какому-то хобби или классу, например
«автомобилист» или «любит спорт». Скоры получаются на основе расчетов ML-модели,
так как мы точно не знаем про всех клиентов, есть ли у них автомобили. Модели сделаны
таким образом, чтобы с максимальной точностью угадывать данные про клиента.
Средняя сумма покупки и обороты измеряются в условных денежных единицах.

### Исследуем, насколько долго пользователи пользуются городом

```python
# Посчитаем число дней между последней и первой транзакцией в городе
df['fuel_delta_days'] = (df['fuel_last_order_date'] - df['fuel_first_order_date']).dt.days.fillna(0)
df['supermarket_delta_days'] = (df['supermarket_last_order_date'] - df['supermarket_first_order_date']).dt.days.fillna(0)
df['afisha_delta_days'] = (df['afisha_last_order_date'] - df['afisha_first_order_date']).dt.days.fillna(0)
```

```python
def func_for_max_day(x, y):
    if x >= y:
        return x
    else:
        return y

df['days_in_gorod'] = df.apply(lambda row: func_for_max_day(row['fuel_delta_days'], row['supermarket_delta_days']), axis=1) 
df['days_in_gorod'] = df.apply(lambda row: func_for_max_day(row['supermarket_delta_days'], row['afisha_delta_days']), axis=1)  
```

```python
min_days_in_gorod = df['days_in_gorod'].min()
max_days_in_gorod = df['days_in_gorod'].max()
```

```python
# Визуализируем
plt.subplots(figsize=(10, 6))

plt.hist(df['days_in_gorod'], bins = 10, alpha = 0.5)

plt.title('Распределения пользователей по дням пользования сервисом Город')
plt.xlabel('Дни пользования сервисом Город')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

**Вывод:** Большая часть пользователей сервиса Город пользуется им менее года.

## Исследование распределния активных пользователей (используют Город) по основным категориям

```python
# Выделим активных пользователей
user_and_category = basic_df[df['gorod_client_flg'] == 1][['party_rk', 'fuel_n_orders', 'supermarket_n_orders', 'afisha_n_orders']]
user_and_category.head(5)
```

```python
# Выделяем группы категорий
columns = ['fuel_n_orders', 'supermarket_n_orders', 'afisha_n_orders']
index = [1, 10, 100]
new_columns = ['fuel', 'supermarket', 'afisha']

for column, i, new_column in zip(columns, index, new_columns):

    def func_for_category(x):
        if x > 0:
            return i
        else:
            return 0

    user_and_category[new_column] = user_and_category[column].apply(func_for_category)   
```

```python
user_and_category.head(5)
```

```python
# Создаем комбинированный категориальный признак
user_and_category['sum_category'] = user_and_category['fuel'] + user_and_category['supermarket'] + user_and_category['afisha']
```

```python
# Называем комбинированный категориальный признак
def func_sum_category_name(x):
        if x == 1:
            return 'Топливо'
        elif x == 10:
            return 'Продукты'
        elif x == 11:
            return 'Топливо и Продукты'
        elif x == 100:
            return 'Афиша'
        elif x == 101:
            return 'Топливо и Афиша'
        elif x == 110:
            return 'Продукты и Афиша'
        else:
            return 'Топливо, Продукты и Афиша'
            
user_and_category['sum_category_name'] = user_and_category['sum_category'].apply(func_sum_category_name) 
```

```python
users_in_sum_category = user_and_category.groupby('sum_category_name')['party_rk'].count().reset_index(name = 'count_users')
users_in_sum_category['count_users'] = round(users_in_sum_category['count_users'].div(len(df[df['gorod_client_flg'] == 1]))*100, 2)
display(users_in_sum_category)
```

```python
# Визуализируем
plt.figure(figsize=(12, 8))

plt.bar(users_in_sum_category['sum_category_name'],
        users_in_sum_category['count_users']
        )

for x,y in zip(users_in_sum_category['sum_category_name'], users_in_sum_category['count_users']):

        label_for_text = round(y, 2)

        plt.annotate(label_for_text, 
                     (x,y), 
                     textcoords="offset points", 
                     xytext=(0,10),
                     ha='center')

plt.title('Распределение пользователей по основным категориям') 
plt.ylabel('Количество пользователей')
plt.xticks(rotation = 45)
plt.show()
```

**Вывод:** Самый большой процент пользователей - 62.22% пользуется только разделом Топливо, второе место делят разделы Афиша и Продукты с одинаковым показателем примерно 11%. Последние два места занимают комбинации Продукты и Афиша, а также Топливо, Продукты и Афиша, в среднем 2%.
